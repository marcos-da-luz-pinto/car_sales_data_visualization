### Importar bibliotecas

```python
# data
import pandas as pd

# data visualization
import plotly.express as px
```

### Carregar o dataset e visualiz√°-lo

```python
# carregar o dataset
caminho_arquivo = '/content/car_sales_data.csv'
car_df= pd.read_csv(caminho_arquivo)

# visualiz√°-lo
car_df.head()
```

```python
# entender o tamanho do dataset e caracter√≠sticas das colunas
car_df.info()
```

### Explorar o dataset

##### 1. Qual fabricante (Manufacturer) ret√©m melhor o pre√ßo em rela√ß√£o √† milhas rodadas (Mileage)?

```python
manufacturer_price_mileage = px.scatter(
    car_df,
    x= 'Mileage',
    y= 'Price',
    color= 'Manufacturer',
    title= 'Deprecia√ß√£o (Milhas x Pre√ßo) por fabricante'
)

manufacturer_price_mileage.show()
```

###### INSIGHTS
- Carros de marcas de luxo, como Porsche e BMW, ret√©m melhor o pre√ßo em rela√ß√£o as milhas rodadas;
- Os carros Toyota t√™m uma reten√ß√£o de valor moderada;
- VW e Ford dominam o mercado de baixo valor.

##### 2. Qual o impacto isolado da Idade (Year of manufacture) e da Quilometragem (Mileage) sobre o Pre√ßo? Qual dos dois fatores tem a correla√ß√£o mais forte com o pre√ßo?

```python
# Year of manufacture x Price
corr = car_df.corr(numeric_only=True)
corr

corr_year_price = px.imshow(
    corr,
    text_auto='.2f',
    title='Mapa de calor das correla√ß√µes'
)

corr_year_price.show()
```

###### INSIGHTS
- A correla√ß√£o entre a vari√°vel 'Ano de fabrica√ß√£o' e 'Pre√ßo' √© positiva forte  (+0,71)
- A correla√ßao entre a vari√°vel 'Milhas' e 'Pre√ßo' √© negativa forte (-0,63)
- Ou seja, se voc√™ tem um carro novo por√©m com altas milhas rodadas, a for√ßa de +0,71 (ano) tentar√° manter o pre√ßo alto, enquanto a for√ßa de -0,63 (milhas) tentar√° abaixar o pre√ßo

##### 3. Como o tamanho do motor (Engine size) e o tipo de combust√≠vel (Fuel type) impactam o valor mediano e a varia√ß√£o de pre√ßo (Price) em carros novos (Year of manufacture > 2018)?

```python
filter_year = car_df[car_df['Year of manufacture']>2018]

car_blox = px.box(
    filter_year,
    x= 'Engine size',
    y= 'Price',
    color= 'Fuel type'
)

car_blox.show()
```

###### INSIGHTS
- Em motores menores, n√£o h√° grandes diverg√™ncias entre medianas e varia√ß√µes de pre√ßo em carros com diferentes tipos de combust√≠vel
- Motores acima de 3L s√£o dominados por carros a gasolina e diesel
- O mercado de carros mais caros com motor potente √©, em sua maioria, Gasolina.
