### Importar bibliotecas

```python
# data
import pandas as pd

# data visualization
import plotly.express as px
```

### Carregar o dataset e visualizá-lo

```python
# carregar o dataset
caminho_arquivo = '/content/car_sales_data.csv'
car_df= pd.read_csv(caminho_arquivo)

# visualizá-lo
car_df.head()
```

```python
# entender o tamanho do dataset e características das colunas
car_df.info()
```

```python
# verificar a existência de dados ausentes (NaN)
car_df.isna().sum()
```

### Explorar o dataset

##### 1. Qual fabricante (Manufacturer) retém melhor o preço em relação à milhas rodadas (Mileage)?

```python
manufacturer_price_mileage = px.scatter(
    car_df,
    x= 'Mileage',
    y= 'Price',
    color= 'Manufacturer',
    title= 'Depreciação (Milhas x Preço) por fabricante'
)

manufacturer_price_mileage.show()
```

###### INSIGHTS
- Carros de marcas de luxo, como Porsche e BMW, retém melhor o preço em relação as milhas rodadas;
- Os carros Toyota têm uma retenção de valor moderada;
- VW e Ford dominam o mercado de baixo valor.

##### 2. Qual o impacto isolado da Idade (Year of manufacture) e da Quilometragem (Mileage) sobre o Preço? Qual dos dois fatores tem a correlação mais forte com o preço?

```python
# Year of manufacture x Price
corr = car_df.corr(numeric_only=True)
corr

corr_year_price = px.imshow(
    corr,
    text_auto='.2f',
    title='Mapa de calor das correlações'
)

corr_year_price.show()
```

###### INSIGHTS
- A correlação entre a variável 'Ano de fabricação' e 'Preço' é positiva forte  (+0,71)
- A correlaçao entre a variável 'Milhas' e 'Preço' é negativa forte (-0,63)
- Ou seja, se você tem um carro novo porém com altas milhas rodadas, a força de +0,71 (ano) tentará manter o preço alto, enquanto a força de -0,63 (milhas) tentará abaixar o preço

##### 3. Como o tamanho do motor (Engine size) e o tipo de combustível (Fuel type) impactam o valor mediano e a variação de preço (Price) em carros novos (Year of manufacture > 2018)?

```python
filter_year = car_df[car_df['Year of manufacture']>2018]

car_blox = px.box(
    filter_year,
    x= 'Engine size',
    y= 'Price',
    color= 'Fuel type'
)

car_blox.show()
```

###### INSIGHTS
- Em motores menores, não há grandes divergências entre medianas e variações de preço em carros com diferentes tipos de combustível
- Motores acima de 3L são dominados por carros a gasolina e diesel
- O mercado de carros mais caros com motor potente é, em sua maioria, Gasolina.
